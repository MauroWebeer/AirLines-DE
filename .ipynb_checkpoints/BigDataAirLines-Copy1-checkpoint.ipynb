```python
from glob import glob 
import pandas as pd
import unidecode
import requests
import json
import re
import os
```

```python
def snake_case(string):
    string = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', string)
    return unidecode.unidecode(re.sub('([a-z0-9])([A-Z])', r'\1_\2', string).lower())
```

```python
airlines_json_file = r'.\VRA\VRA_20211.json'
airlines_json = pd.read_json(airlines_json_file, encoding='utf-8-sig')
airlines_json.head()
```

```python
lista_nomes = airlines_json.columns.values.tolist()
lista_nomes
```

```python
nova_lista = []
for i in lista_nomes:
    nova_lista.append(snake_case(i))
print(nova_lista)
```

```python
file_list = glob(os.path.join('.\VRA', "*.json"))
```

```python
dfs = []
for file in file_list:
    with open(file,encoding='utf-8-sig') as f:
        json_data = pd.json_normalize(json.loads(f.read()))
        json_data['site'] = file.rsplit("\\", 1)[-1]
    dfs.append(json_data)
df = pd.concat(dfs)
```
