```python
import glob 
import pandas as pd
import unidecode
import requests
import json
from re import sub
import os
```

```python
#Iremos, primeiro, ler a base de voos criada anteriormente para retirada codigos icao que serao
#utilizados na chamada da API. 

#Passamos o dtype para algumas colunas pois estava ocorrendo um erro de multiplos data types na 
#mesma coluna
df = pd.read_csv('./airlines_VRA.csv', dtype={'icao_empresa_aerea':'str','numero_voo': 'str','codigo_autorizacao':'str'})

#Iremos guardar todos os codigos que aparecem em origem e destino numa lista
lista_icao = []
for i in df.icao_aerodromo_origem.unique():
    lista_icao.append(i)
for i in df.icao_aerodromo_destino.unique():
    lista_icao.append(i)

    
#Com auxilio do myset retiramos todas as repetições de códigos
myset = set(lista_icao)
icao_list = list(myset)
```

```python
#Ao invés da url = 'https://rapidapi.com/Active-api/api/airport-info/' iremos utilizar
#a url sugerida na documentação da API

#Usaremos esta lista para guardar todas as respostas recebidas da API
json_list = []

url = "https://airport-info.p.rapidapi.com/airport"
headers = {
 "X-RapidAPI-Key": "10557bd5bcmsh02ce51cdb270c15p1a5db9jsn60482611d5ea",
 "X-RapidAPI-Host": "airport-info.p.rapidapi.com"
}

#Passaremos como parametros, alem da url e do hearder acima, a lista de icao tomada acima
for querystring in icao_list:
    response = requests.request("GET", url, headers=headers, params={'icao':querystring})
    json_data = json.loads(response.text)
    json_list.append(json_data)

#Transformaremos o resultado encontrado em um dataframe
jsonString = json.dumps(json_list)
json_test = json.loads(jsonString)
df = pd.json_normalize(json_test)
df
```

```python
df.to_csv('airlines_aerodromos_VRA.csv', index=False, encoding='utf-8-sig')
```
